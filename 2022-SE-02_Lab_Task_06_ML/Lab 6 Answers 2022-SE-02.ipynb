```python
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(df=df)
plt.xticks(rotation=90)
plt.title('Boxplot for Outlier Detection')
plt.show()
```

```python
categorical_features = ['Time_of_Day', 'Day_of_Week', 'Traffic_Conditions', 'Weather']
label_encoders = {}
```

```python
for col in categorical_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
X = df.drop(columns=['Trip_Price'])
y = df['Trip_Price']
```

```python
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1
X = X[~((X < (Q1 - 1.5 * IQR)) | (X > (Q3 + 1.5 * IQR))).any(axis=1)]
y = y.loc[X.index]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
```

```python
print("RMSE:", mean_squared_error(y_test, y_pred_lr))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lr)))
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=y_test, y=y_pred_lr, label='Linear Regression')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Model Predictions vs Actual Prices')
plt.legend()
plt.show()
```
